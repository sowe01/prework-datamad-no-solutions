```python
#Rock Paper Scissors

# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

gestures=['stone','paper','scissors']


#3. Create a variable called n_rounds to store the maximum number of rounds to play in a game

n_rounds=int(input("introduce un numero impar de rondas maximas que quieres jugar: "))
if n_rounds %2 ==0:
   n_rounds=int(input("introduce un numero impar : "))
    
# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
import math
rounds_to_win= math.ceil(n_rounds/2)


#Create two variables to store the number of rounds that the computer and the player have won. 
#Call these variables cpu_score and player_score.

cpu_score=0
player_score=0

#6. Define a function that randomly returns one of the 3 gesture options
def pc_option():
    play_cpu=random.choice(gestures)
    return play_cpu

#7. Define a function that asks the player which is the gesture he or she wants to show: 
#'rock', 'paper' or 'scissors'


def player_option():
    
    valido=False
    
    while not valido:
        opcion= input("introduce una opcion, stone, paper, scissors: ")
        if opcion in gestures:
            valido=True
        
    return opcion


#8. Define a function that checks who won a round.

def check_result (pc,player):
    if pc == player:
        return 0
    if pc=="stone":
        return 1 if player =="scissors" else 2
    if pc=="paper":
        return 1 if player =="stone" else 2
    if pc=="scissors":  
        return 1 if player == "paper" else 2
    
#9. Define a function that prints the choice of the computer, the choice of the player 
#and a message that announces who won the current round

def round():
    global cpu_score, player_score
    pc= pc_option()
    player= player_option() 
    result=check_result(pc,player)
    print("el pc elige: ", pc, ", el jugador elige: ", player)
    if result==0:
        print("el resultado es un empate")
    elif result==1:
        print("el pc gana")
        cpu_score +=1
    else: 
        print("gana el jugador")
        player_score+=1
    print ("pc score: ",cpu_score, "player score: ", player_score)


        
        
# Create two variables that accumulate the wins of each participant     

while cpu_score<rounds_to_win and player_score<rounds_to_win:
    round()

if cpu_score>player_score:
    print("gana el pc")
else:
    print("gana el jugador")
```
